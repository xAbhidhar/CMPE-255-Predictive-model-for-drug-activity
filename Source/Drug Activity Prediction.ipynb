```python
#Importing the required libraries
import pandas as pd
import scipy
import numpy as np
from sklearn.ensemble import RandomForestClassifier as RFC
#counter decalration
number=0

#Reading the testing and training data and saving it in dataframe after splitting by new lines
with open("test.dat","r") as f:
    test_data_frame = pd.DataFrame(l.split("\t") for l in f) 
with open("train.dat","r") as train_data:
    data_frame_train = pd.DataFrame(l.split("\t") for l in train_data)
test_data_frame.columns = ["attributes"]  
test_data_frame_true = pd.DataFrame(test_data_frame["attributes"].str.split("\n")) 
data_frame_train.columns = ["values","attributes"]  
data_frame_attributes = pd.DataFrame(data_frame_train["attributes"].str.split("\n"))
data_frame_values = pd.DataFrame(data_frame_train["values"].str.split("\n"))
data_attribute = data_frame_train["attributes"]
data_values = data_frame_train["values"]

#Creating the Random Forest Classifier
#RFC_classifier = RFC(n_estimators=12)
#RFC_classifier = RFC(n_estimators=6)
#RFC_classifier = RFC(n_estimators=10)
#RFC_classifier = RFC(n_estimators=2)
RFC_classifier = RFC(n_estimators=25)
RFC_data_values = data_values.as_matrix()
RFC_data_values = [int(x) for x in RFC_data_values]

#Creating an array for the attributes and creating input data for classifier
np_array = np.empty(800, dtype=np.object)
for f in data_frame_attributes["attributes"]:
        for q in f:
            s = q.rstrip(" ")
            c = s.rstrip(",")
            s1 = c.split(" ")
            array_attribute =[]
            for i in s1:
                if(len(i)>0 and int(i)<=100000):
                    array_attribute.append(i)
            if(len(array_attribute)>1):
                array_attribute = [int(i) for i in array_attribute]
                array_attributes = np.asarray(array_attribute)
            np_array[number] = array_attributes 
        number += 1           
RPC_model = [[]*800 for p in range(800)]
model_array = np.asarray(RFC_data_values)
for x in range(800):
    temp_list = [0] * 100000
    for q in np_array[x]:
        temp_list[q-1] = 1     
    temp_list_array = np.asarray(temp_list)
    RPC_model[x] = temp_list_array
    x += 1 

#Loading the data from the RPC_model and converting into array
RPC_model_array = np.array(RPC_model)

#Training the model with the the fit method from RFC
RFC_classifier.fit(RPC_model_array,model_array)

#Writing the vlaues to the result.dat 
result_file = open("result.dat", 'w') 
for z in test_data_frame_true["attributes"]:
    #print "Inside the loop 1"
        for q in z:
            #print "Inside the loop 2"
            space_strip = q.rstrip(" ")
            other = space_strip.rstrip(",")
            other = space_strip.rstrip("\n")
            split_final = other.split(" ")
            attributes_array =[]
            for x in split_final:
                #print "Inside the loop 3"
                if(len(x)>0 and int(x)<=100000):
                    #print "Inside the loop 4"
                    attributes_array.append(x)
            if(len(attributes_array)>1):
                attributes_array = [int(x) for x in attributes_array]
                attributes_array_true = np.asarray(attributes_array)  
        temp_list = [0] * 100000
        for q in attributes_array_true:
            #print "Inside the loop 2A"
            temp_list[q-1] = 1
        final_array = np.asarray(temp_list)
        if(len(final_array)>1):
            #print "Inside the loop 2B - Yay writing!!!"
            output_result = RFC_classifier.predict(final_array)
            if(output_result[0] == 1):
                result_file.write("1\n")
            if(output_result[0] == 0):
                result_file.write("0\n")

#Closing the output file
result_file.close()

#print "File written lets check output"
```
